# EDA

### 1. User Demographics

```python
# Setting up subplots to visualize fraudulent activities across multiple demographic variables
fig, axs = plt.subplots(3, 2, figsize=(15, 12))

# Fraudulent activities by Age
sns.histplot(df[df['IsFraud'] == 1]['Age'], bins=10, kde=False, ax=axs[0, 0], color='blue')
axs[0, 0].set_title('Fraudulent Activities by Age')
axs[0, 0].set_xlabel('Age')
axs[0, 0].set_ylabel('Count')

# Fraudulent activities by Gender (Male)
sns.barplot(x=df[df['IsFraud'] == 1]['Gender_Male'].value_counts().index,
            y=df[df['IsFraud'] == 1]['Gender_Male'].value_counts().values, ax=axs[0, 1], palette='coolwarm')
axs[0, 1].set_title('Fraudulent Activities by Gender (Male)')
axs[0, 1].set_xlabel('Gender')
axs[0, 1].set_ylabel('Count')
axs[0, 1].set_xticklabels(['Female', 'Male'])

# Fraudulent activities by Occupation
sns.countplot(y=df[df['IsFraud'] == 1]['Occupation'], ax=axs[1, 0], palette='coolwarm', order=df['Occupation'].value_counts().index)
axs[1, 0].set_title('Fraudulent Activities by Occupation')
axs[1, 0].set_xlabel('Count')
axs[1, 0].set_ylabel('Occupation')

# Fraudulent activities by Education Level
sns.countplot(y=df[df['IsFraud'] == 1]['EducationLevel'], ax=axs[1, 1], palette='coolwarm', order=df['EducationLevel'].value_counts().index)
axs[1, 1].set_title('Fraudulent Activities by Education Level')
axs[1, 1].set_xlabel('Count')
axs[1, 1].set_ylabel('Education Level')

# Fraudulent activities by Marital Status
sns.countplot(y=df[df['IsFraud'] == 1]['MaritalStatus'], ax=axs[2, 0], palette='coolwarm', order=df['MaritalStatus'].value_counts().index)
axs[2, 0].set_title('Fraudulent Activities by Marital Status')
axs[2, 0].set_xlabel('Count')
axs[2, 0].set_ylabel('Marital Status')

# Fraudulent activities by Number of Dependents
sns.countplot(x=df[df['IsFraud'] == 1]['NumDependents'], ax=axs[2, 1], palette='coolwarm')
axs[2, 1].set_title('Fraudulent Activities by Number of Dependents')
axs[2, 1].set_xlabel('Number of Dependents')
axs[2, 1].set_ylabel('Count')

plt.tight_layout()
plt.show()
```

```python
# Segmenting 'Age' into categories (bins) and visualizing the fraud count
df['AgeGroup'] = pd.cut(df['Age'], bins=[18, 25, 35, 45, 55, 65, 75], labels=['18-25', '25-35', '35-45', '45-55', '55-65', '65-75'])

# Plotting the segmented age groups with fraud count
plt.figure(figsize=(10, 6))
sns.countplot(x='AgeGroup', hue='IsFraud', data=df, palette='coolwarm')
plt.title('Fraudulent Activities by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Count')
plt.tight_layout()
plt.show()
```

```python
# Plotting crosstab visualizations for the 5 demographic variables using the coolwarm palette

plt.figure(figsize=(18, 12))

# Crosstab for Gender
plt.subplot(2, 3, 1)
gender_crosstab = pd.crosstab(df['Gender_Male'], df['IsFraud'], normalize='index')
gender_crosstab.plot(kind='bar', stacked=True, color=sns.color_palette('coolwarm', 2), ax=plt.gca())
plt.title('Fraud by Gender')
plt.xlabel('Gender (0=Female, 1=Male)')
plt.ylabel('Proportion')

# Crosstab for Occupation
plt.subplot(2, 3, 2)
occupation_crosstab = pd.crosstab(df['Occupation'], df['IsFraud'], normalize='index')
occupation_crosstab.plot(kind='bar', stacked=True, color=sns.color_palette('coolwarm', 2), ax=plt.gca())
plt.title('Fraud by Occupation')
plt.xlabel('Occupation')
plt.ylabel('Proportion')

# Crosstab for Education Level
plt.subplot(2, 3, 3)
education_crosstab = pd.crosstab(df['EducationLevel'], df['IsFraud'], normalize='index')
education_crosstab.plot(kind='bar', stacked=True, color=sns.color_palette('coolwarm', 2), ax=plt.gca())
plt.title('Fraud by Education Level')
plt.xlabel('Education Level')
plt.ylabel('Proportion')

# Crosstab for Marital Status
plt.subplot(2, 3, 4)
marital_crosstab = pd.crosstab(df['MaritalStatus'], df['IsFraud'], normalize='index')
marital_crosstab.plot(kind='bar', stacked=True, color=sns.color_palette('coolwarm', 2), ax=plt.gca())
plt.title('Fraud by Marital Status')
plt.xlabel('Marital Status')
plt.ylabel('Proportion')

# Crosstab for Number of Dependents
plt.subplot(2, 3, 5)
dependents_crosstab = pd.crosstab(df['NumDependents'], df['IsFraud'], normalize='index')
dependents_crosstab.plot(kind='bar', stacked=True, color=sns.color_palette('coolwarm', 2), ax=plt.gca())
plt.title('Fraud by Number of Dependents')
plt.xlabel('Number of Dependents')
plt.ylabel('Proportion')

plt.tight_layout()
plt.show()
```

```python
# Performing crosstab analysis for the 6 User Demographics variables
demographics_vars = ['Gender_Male', 'Occupation', 'EducationLevel', 'MaritalStatus', 'NumDependents']

# Creating a dictionary to hold crosstab results for each demographic variable with respect to fraud
crosstab_results = {}
for var in demographics_vars:
    crosstab_results[var] = pd.crosstab(df[var], df['IsFraud'], normalize='index')

# Combining the crosstab results into a single dataframe for analysis
combined_crosstab = pd.concat(crosstab_results)

# Displaying the top crosstabs for review
combined_crosstab
```

### Key Takeaways:

- Age: Fraudulent activities are distributed across various age groups, with a concentration around younger and middle-aged individuals.
- Gender (Male): There is a slightly higher count of fraudulent activities involving males than females.
- Occupation: Certain occupations, such as professionals and students, are associated with more fraud cases.
- Education Level: Fraudulent activities are prevalent across different education levels, with some categories like 'Bachelor' having more instances.
- Marital Status: Fraud is more common among married individuals, followed by single and widowed.
- Number of Dependents: Fraudulent activities are fairly spread across different numbers of dependents, though those with 3-4 dependents show higher counts.

**Cross tab analysis**


- Students and individuals with more dependents show higher rates of fraud.
- Widowed and single individuals also exhibit a slightly higher tendency toward fraud.
- Education level and occupational stability seem to lower the likelihood of fraud.

### 2. Transaction

#### 2.1 Time-related Information Visualization

```python

plt.figure(figsize=(15, 10))

# Fraudulent activities by Weekday
plt.subplot(2, 2, 1)
sns.countplot(x='Weekday', data=df[df['IsFraud'] == 1], palette='coolwarm')
plt.title('Fraudulent Activities by Weekday')

# Fraudulent activities by Time of Day
plt.subplot(2, 2, 2)
sns.countplot(x='TimeofDay', data=df[df['IsFraud'] == 1], palette='coolwarm')
plt.title('Fraudulent Activities by Time of Day')

# Fraudulent activities by Transaction Time (Hour)
plt.subplot(2, 2, 3)
df['TransactionHour'] = pd.to_datetime(df['TransactionTime']).dt.hour
sns.histplot(df[df['IsFraud'] == 1]['TransactionHour'], bins=24, kde=False, color='blue')
plt.title('Fraudulent Activities by Transaction Time (Hour)')
plt.xlabel('Hour of Day')

plt.tight_layout()
plt.show()
```

```python
# Convert TransactionDate to proper datetime format with year-month-date
df['TransactionDate'] = pd.to_datetime(df['TransactionDate'], format='%Y/%m/%d')

# Extract the year and month for analysis
df['YearMonth'] = df['TransactionDate'].dt.to_period('M')

# Plot fraudulent activities over time by Year-Month
plt.figure(figsize=(10, 6))
sns.countplot(x='YearMonth', hue='IsFraud', data=df, palette='coolwarm')
plt.title('Fraudulent Activities Over Time (Year-Month)')
plt.xlabel('Year-Month')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Get the count of fraudulent and non-fraudulent activities per month
monthly_fraud_counts = df.groupby(['YearMonth', 'IsFraud']).size().unstack(fill_value=0)

# Calculate the percentage of fraudulent activities each month
monthly_fraud_counts['FraudPercentage'] = (monthly_fraud_counts[1] / 
                                           (monthly_fraud_counts[0] + monthly_fraud_counts[1])) * 100
monthly_fraud_counts['FraudPercentage']
```

#### 3.3 Segmenting into groups

```python
# Merge all the segmentation and plotting code into one block

# Segment transaction amounts into categories (Low, Medium, High, Very High)
df['TransactionAmountSegment'] = pd.qcut(df['TransactionAmount_AUD'], q=4, labels=['Low', 'Medium', 'High', 'Very High'])

# Segment gift transaction amounts into categories (Low, Medium, High, Very High)
df['GiftsTransactionSegment'] = pd.qcut(df['GiftsTransaction_AUD'], q=4, labels=['Low', 'Medium', 'High', 'Very High'])

# Segment income amounts into categories (Low, Medium, High, Very High)
df['IncomeSegment'] = pd.qcut(df['Income_AUD'], q=4, labels=['Low', 'Medium', 'High', 'Very High'])

# Segment expenditure amounts into categories (Low, Medium, High, Very High)
df['SpendingGroup'] = pd.qcut(df['Expenditure_AUD'], q=4, labels=['Low', 'Medium', 'High', 'Very High'])

# Plot the segmented transaction amounts, gifts transaction amounts, income amounts, and expenditure with fraud count
plt.figure(figsize=(18, 12))

# Transaction Amount Segment
plt.subplot(2, 2, 1)
sns.countplot(x='TransactionAmountSegment', hue='IsFraud', data=df, palette='coolwarm')
plt.title('Fraudulent Activities by Transaction Amount Segment')
plt.xlabel('Transaction Amount Segment')
plt.ylabel('Count')

# Gifts Transaction Segment
plt.subplot(2, 2, 2)
sns.countplot(x='GiftsTransactionSegment', hue='IsFraud', data=df, palette='coolwarm')
plt.title('Fraudulent Activities by Gifts Transaction Amount Segment')
plt.xlabel('Gifts Transaction Amount Segment')
plt.ylabel('Count')

# Income Amount Segment
plt.subplot(2, 2, 3)
sns.countplot(x='IncomeSegment', hue='IsFraud', data=df, palette='coolwarm')
plt.title('Fraudulent Activities by Income Amount Segment')
plt.xlabel('Income Amount Segment')
plt.ylabel('Count')

# Spending Group (Expenditure)
plt.subplot(2, 2, 4)
sns.countplot(x='SpendingGroup', hue='IsFraud', data=df, palette='coolwarm')
plt.title('Fraudulent Activities by Spending Group')
plt.xlabel('Spending Group')
plt.ylabel('Count')

plt.tight_layout()
plt.show()
```

##### General Observations:
- Across all categories, fraudulent activity is more prevalent in the Low and Medium value segments, suggesting that fraud detection systems should pay more attention to smaller transactions, gifts, and expenditures.
- Higher value transactions may be subject to more stringent checks, leading to fewer fraudulent activities in those segments.
